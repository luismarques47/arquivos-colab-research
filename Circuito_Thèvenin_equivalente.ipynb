```python
import numpy as np
A = np.array([[1, -1,0,0],[20, 10,-10,0],[0, -10,40,-20],[0, 0,-20,40]])
B = np.array([5,30,0,20])
A_inversa = np.linalg.inv(A)
X = np.dot(A_inversa,B)
print('O valor de i1 = {:.3f} A'.format(X[0]))
print('O valor de i2 = {:.3f} A'.format(X[1]))
print('O valor de i3 = {:.3f} A'.format(X[2]))
print('O valor de i4 = {:.3f} A'.format(X[3]))
```

```python
Va = 30 - 10*X[0]
print('O valor de Va = {:.3f} V'.format(Va))
```

```python
Vb = 10*X[2]
print('O valor de Vb = {:.3f} V'.format(Vb))
```

```python
Vth = Va-Vb
print('O valor de Vth = {:.3f} V'.format(Va - Vb))
```
