```python
import sympy as sp
from sympy import I,Heaviside


Vm , t, w, a= sp.symbols('Vm t w a')

f = sp.Function('f')
f= Vm
f
h = f*(Heaviside(t+a)-Heaviside(t-a))
sp.plot(h.subs(((a,0.5),(f,1))) ,(t,-3,3));
```

# ***Transformada de Fourier***

```python
X = sp.integrate(exp(-I*w*t),(t,-0.5,0.5))
X
```

# **GRÁFICO PARA X(W)**

```python
sp.plot(X,(w,-50,50))
```
