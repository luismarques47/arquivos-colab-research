# -0,5Vo+iaux+Vo/20  = 0 
# 20*io+10*io+1 = 0 
# io = Vo/20 

```python
import numpy as np
```

```python
A = np.array([[(-0.5+0.05), 1, 0],[0, 0, 30],[0.05, 0, -1]])
B = np.array([0,-1,0])
A_inversa = np.linalg.inv(A)
X = np.dot(A_inversa,B)
print('O valor de Vo = {:.3f} V'.format(X[0]))
print('O valor de iaux = {:.3f} A'.format(X[1]))
print('O valor de io = {:.3f} A'.format(X[2]))
```
