# -70 +30k*io+4*Vo  = 0 
# 30k*io+4*Vo = 70 
# V0 = 10k*io
# 10k*io - V0 = 0 

```python
import numpy as np
A = np.array([[30000, 4],[10000, -1]])
B = np.array([70,0])
A_inversa = np.linalg.inv(A)
X = np.dot(A_inversa,B)
print('O valor de Io = {:.3f} A'.format(X[0]))
print('O valor de Vo = {:.3f} V'.format(X[1]))
```

```python
Vth = 70 - X[1]
print('O valor de Vth = {:.3f} V'.format(Vth))
```

# icc = i1 - i2

# -70 +10k*i1 = 0
# 20k*i2+4V0 = 0
# V0 = 10ki1

# 10000i1 = 70
# 20000i2 + 4V0 = 0
# 10000i1 - V0 = 0

```python
A = np.array([[10000, 0, 0],[0, 20000, 4],[10000, 0, -1]])
B = np.array([70,0,0])
A_inversa = np.linalg.inv(A)
X = np.dot(A_inversa,B)
print('O valor de I1 = {:.3f} A'.format(X[0]))
print('O valor de I2 = {:.3f} A'.format(X[1]))
print('O valor de Vo = {:.3f} V'.format(X[2]))
print('O valor de Icc = {:.3f} A'.format(X[0]-X[1]))
```

```python
print('O valor de Rth = {:.3f} ohms'.format(Vth/(X[0]-X[1])))
```
