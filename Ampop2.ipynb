Sabendo que Vs = 0, determine ***io*** no circuito abaixo.

```python
import sympy as sp
v0 = sp.symbols('v0')
i0 = sp.symbols('i0')
eq1 = sp.Eq(-0.01/4000-v0/8000,0)
V0 = sp.solve(eq1,v0)

# Convert list of items to a string value
delimiter = ' '
string = delimiter.join(map(str, V0))

V0 = float(string)
i0 = V0/2000
print('O valor de i0 = {:.2E} A'.format(i0))
```
