# Exercício
Determine ***vo/vs*** no circuito abaixo.

Captura de Tela 2021-06-29 às 14.50.16.png

```python
import sympy as sp
v0 = sp.symbols('v0')
vs = sp.symbols('vs')
eq1 = sp.Eq(-vs/5000-v0/25000,0)
resultados = sp.solve(eq1,[v0,vs])
resultados
```

```python
type(resultados)
```

```python
print(resultados)
```

```python
print(resultados[0][0])
print(resultados[0][1])
print(resultados[0][0]/resultados[0][1])
```
