```python
import sympy as sp
sp.init_printing()
```

```python
s = sp.Symbol('s')
t = sp.Symbol('t')
v1 = sp.Symbol('v1')
v0 = sp.Symbol('v0')
```

```python
eq1 = sp.Eq(-3/s+v1/(2*s)+v1-v0,0)
eq2 = sp.Eq(v0/2+(v0-10/s)/(2/s)+v0-v1,0)

sp.solve([eq1,eq2],[v1,v0])
```

```python
f = s**2+1.5*s+1.5
sp.solve(f)
```

```python
V0 = (10*s+11)/ ((s+0.75+0.97j) * (s+0.75-0.97j))
V0
```

```python
A = sp.Symbol('A')
B =  sp.Symbol('B')

eq3 = sp.Eq((10*(-0.75+0.97j)+11)/((-0.75+0.97j)+0.75+0.97j),A)
eq4 = sp.Eq((10*(-0.75 - 0.97j)+11)/ ( -0.75 - 0.97j + 0.75  - 0.97j),B)

sp.solve([eq3,eq4],[A,B])
```
