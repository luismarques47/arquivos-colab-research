# Exercício
Determine ***Vo e io*** no circuito abaixo.

Captura de Tela 2021-06-29 às 19.11.40.png

```python
import sympy as sp
v0 = sp.symbols('v0')
i0 = sp.symbols('i0')
vp = sp.symbols('vp')
eq1 = sp.Eq(-vp+90/100,0)
eq2 = sp.Eq(-i0+(v0-vp)/100000+v0/10000,0)
eq3 = sp.Eq(-vp+(50*v0)/150,0)


sp.solve([eq1,eq2,eq3],[v0,i0,vp])
```
